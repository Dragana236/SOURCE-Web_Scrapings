```python
from selenium import webdriver
from bs4 import BeautifulSoup
import pandas as pd

class Player():
    def init(self):
        self.name=''
        self.link=''
        
def get_player_list():
    driver=webdriver.PhantomJS(executable_path=r'C:\Users\Dragana\Anaconda2\phantomjs-2.1.1-windows\phantomjs-2.1.1-windows\bin\phantomjs.exe')
    url=('http://www.nba.com/players')
    driver.get(url)
    html_doc=driver.page_source
    #print driver.page_source
    soup=BeautifulSoup(html_doc,'lxml')
    players=soup.find('div',class_='row players-wrapper')
    
    player_list=[]
    for player in players.find_all('a'):
        full=player.text
        player_name=full.split('#')[0]
        part=player['href']
        player_link='http://www.nba.com'+part
        new_play=Player()
        new_play.name=player_name
        new_play.link=player_link
        player_list.append(new_play)
    for one_player in player_list:
        print one_player.name
        print one_player.link
    driver.quit()
    return player_list
#get_player_list()
```
