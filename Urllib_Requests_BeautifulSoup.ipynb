```python
#Get requests using requests package
#higher-level requests library
#provides a perfect API for making requests
#you don't have to close the connection when using requests
import requests
url='https://www.wikipedia.org/'
r=requests.get(url)    #you are packaging request, sending request and catching response with the single function requests.get()
text=r.text  #apply .text method (text attribute of the object r) to response which returns HTML as a string
```

```python
#Scraping the web in Python
#you got html which is mix of unstructured and structured data
#BeautifulSoup-parse html and extract structured data from HTML, make tag soup beautiful and extract information
#the object created is calles BeautitifulSoup and it has associated method prettify
from bs4 import BeautifulSoup
import requests   #with requests you scrape HTML from a webpage
url='https://www.crummy.com/software/BeautifulSoup/'
r=requests.get(url)
html_doc=r.text
soup=BeautifulSoup(html_doc) #you create BS object from HTML and prettify it
print(soup.prettify())
```
