```python
#if you want to load a file from the web into DataFrame without saving it first locally you can use pd.read_csv() with url as the 1st arg 
import matplotlib.pyplot as plt #for later ploting
import pandas as pd
url='http://archive.ics.uci.edu/ml/machine-learning-databases/wine-quality/winequality-white.csv'
df=pd.read_csv(url,sep=';')
df.head()
```

```python

pd.DataFrame.hist(df.ix[:,0:1])
plt.xlabel('fixed acidity')
plt.ylabel('count')
plt.show()
```

```python
pd.DataFrame.hist(df.iloc[:,0:1])
plt.xlabel('fixed acidity')
plt.ylabel('count')
plt.show()
```
